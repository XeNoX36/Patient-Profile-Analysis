```python
import pandas as pd
```

**Data Cleaning and Exploration**

```python
df = pd.read_csv(r"c:\Users\USER\Documents\CODES\Python Work to Do\Hospital Analysis\healthcare_dataset.csv")
df
```

*Approximating all floats to 2 decimal places*

```python
pd.set_option('display.float_format', lambda x: '%.2f' % x)
```

*Cleaning Name Column*

```python
df["Name"] = df["Name"].str.title()
df["Name"]
```

*Categorizing the Age column*

```python
df["Age_groups"] = pd.cut(df["Age"], bins=[0, 25, 40, 65, 80, 90], labels=["13-24", "25-39", "40-64", "65-79", "80+"])
df["Age_groups"]
```

*Creating an Admission Duration Column*

```python
df["Date of Admission"] = pd.to_datetime(df["Date of Admission"])
df["Discharge Date"] = pd.to_datetime(df["Discharge Date"])
```

```python
df["Admission_Duration"] = df["Discharge Date"] - df["Date of Admission"]
df
```

```python
df.describe()
```

```python
df[df["Billing Amount"] < 0]
# meaning we have some debtors
```

```python
df["Billing Amount"] = df["Billing Amount"].abs()
df[df["Billing Amount"] < 0]
```

```python
df.drop(["Hospital", "Room Number"], axis=1, inplace=True)
```

*Adding Year, Month and day columns*

```python
df["Year"] = df["Date of Admission"].dt.year
df["Month"] = df["Date of Admission"].dt.strftime("%B")
df["Day"] = df["Date of Admission"].dt.day_name()
df
```

```python
df["Admission_Duration"] = df["Admission_Duration"].astype(str)
df["Admission_Duration"] = df["Admission_Duration"].str.replace(" days", "")
df["Admission_Duration"] = df["Admission_Duration"].astype(int)
df
```

*Printing-out the processed dataset*

```python
df.to_csv(r"c:\Users\USER\Documents\CODES\Python Work to Do\Hospital Analysis\Cleaned_healthcare.csv", index=True)
```
